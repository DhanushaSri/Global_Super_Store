### 3. Data Analysis

```python
#DISCOUNT IMPACT ANALYSIS
correlation_sales = df[['Shipping Cost', 'Sales']].corr().iloc[0, 1]
correlation_profit = df[['Shipping Cost', 'Profit']].corr().iloc[0, 1]

print(f"Correlation between Shipping Cost and Sales: {correlation_sales:.2f}")
print(f"Correlation between Shipping Cost and Profit: {correlation_profit:.2f}")
```

```python
#ORDER PRIORITY ANALYSIS
priority_summary = df.groupby('Order Priority').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
print(priority_summary)
```

```python
#TOTAL PROFIT
profit = data.groupby('category')['profit'].sum().sort_values(ascending=False)
profit
```

```python
# TOTAL SALES AND PROFIT BY CATEGORY
category_summary = df.groupby('Category').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
print(category_summary)
```

```python
#SALES PERFORMANCE BY REGION
sales_by_region = df.groupby('Region').agg({'Sales': 'sum'}).reset_index()
print(sales_by_region)
```

```python
#TOTAL SALES
sales = data.groupby('category')['sales'].sum().sort_values(ascending=False)
sales
```

```python
#CUSTOMER SEGMENTATION ANALYSIS
customer_summary = df.groupby('Customer ID').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
bins = [0, 500, 1000, 2000]
labels = ['Low', 'Medium', 'High']
customer_summary['Sales Segment'] = pd.cut(customer_summary['Sales'], bins=bins, labels=labels)
print(customer_summary)
```

```python
#TOP PRODUCTS BY SALES AND PROFIT
top_products_sales = df.sort_values(by='Sales', ascending=False).head(5)
top_products_profit = df.sort_values(by='Profit', ascending=False).head(5)

print("Top 5 Products by Sales:")
print(top_products_sales)
print("\nTop 5 Products by Profit:")
print(top_products_profit)
```

```python
# PROFIT MARGIN
sales_category = data.groupby(['category', 'sub-category'], as_index=False)[['sales', 'profit']].sum()
sales_category['profit_margin'] = sales_category['profit'] / sales_category['sales']
sales_category =  sales_category.sort_values(by='profit_margin', ascending=False)
sales_category 
```

```python
#ANALYSING MONTHLY TRENDS
df = pd.DataFrame(data)

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
monthly_trends = df.groupby(['Year', 'Month']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
monthly_trends['Date'] = pd.to_datetime(monthly_trends[['Year', 'Month']].assign(DAY=1))

print(monthly_trends)
```

```python
# Aggregate total Sales and Profit by Customer ID
cus_summary = df.groupby('Customer ID').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

# Calculate average Sales and Profit
avg_s_per_cus = cus_summary['Sales'].mean()
avg_p_per_cus = cus_summary['Profit'].mean()

# Calculate average profit margin
avg_profit_margin = avg_p_per_cus / avg_s_per_cus

# Calculate Customer Lifetime Value (CLV)
clv = avg_s_per_cus * avg_profit_margin

print(f"Average Sales per Customer: ${avg_s_per_cus:.2f}")
print(f"Average Profit per Customer: ${avg_p_per_cus:.2f}")
print(f"Average Profit Margin: {avg_profit_margin:.2f}")
print(f"Customer Lifetime Value (CLV): ${clv:.2f}")
```

### 4. Data Visualization

```python
#BAR CHART FOR TOTAL SALES AND PROFIT BY CATEGORY
sales = data.groupby('category')['sales'].sum()
profit = data.groupby('category')['profit'].sum()

title_font = {'family': 'serif', 'color': 'darkblue', 'weight': 'bold', 'size': 20 }
label_font = {'family': 'serif', 'color': 'darkred', 'weight': 'normal', 'size': 14 }
tick_font = {'family': 'sans-serif', 'color': 'black', 'size': 12 }


plt.figure(figsize=(16, 12))

plt.subplot(1, 2, 1) 
sns.barplot(x=sales.index, y=sales.values, palette='viridis')
plt.title('Total Sales by Category', **title_font)
plt.xlabel('Category', **label_font)
plt.ylabel('Sales', **label_font)
plt.xticks(rotation=45, ha='right', **tick_font)
plt.yticks(**tick_font)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(1, 2, 2)  
sns.barplot(x=profit.index, y=profit.values, palette='viridis')
plt.title('Total Profit by Category', **title_font)
plt.xlabel('Category', **label_font)
plt.ylabel('Profit', **label_font)
plt.xticks(rotation=45, ha='right', **tick_font)
plt.yticks(**tick_font)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# HISTOGRAM FOR DISTRIBUTION OF PROFIT MARGINS BY SAUB CATEGORY 
profit_margin = sales_category['profit_margin']

plt.figure(figsize=(12, 8))
sns.histplot(
    profit_margin,
    bins=20,
    color='blue',
    edgecolor='black',
    kde=True,
    alpha=0.75
)

plt.title('Distribution of Profit Margins by Sub-category', fontsize=20, fontweight='bold')
plt.xlabel('Profit Margin', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)

plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter(1.0))

for p in plt.gca().patches:
    width = p.get_width()
    height = p.get_height()
    plt.gca().annotate(f'{int(height)}', (width, height), ha='center', va='bottom', fontsize=12, color='black')

plt.tight_layout()
plt.show()
```

```python
#DISTRIBUTION OF SALES BY CATEGORY PER SEGMENT
data = {
    'segment': ['Consumer', 'Consumer', 'Consumer', 'Corporate', 'Corporate', 'Corporate', 'Home Office', 'Home Office', 'Home Office'],
    'category': ['Furniture', 'Office Supplies', 'Technology', 'Furniture', 'Office Supplies', 'Technology', 'Furniture', 'Office Supplies', 'Technology'],
    'sales': [400, 300, 250, 450, 200, 300, 500, 350, 200]
}
seg_category = pd.DataFrame(data)

# Pivot the data for the stacked bar chart
pivot_data = seg_category.pivot(index='category', columns='segment', values='sales')

# Plot a stacked bar chart
pivot_data.plot(kind='bar', stacked=True, figsize=(12, 8), color=['#1f77b4', '#ff7f0e', '#2ca02c'])

# Add labels and title
plt.title('Sales by Category per Segment', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.xticks(rotation=45, ha='right')

# Add a legend
plt.legend(title='Segment')

# Show plot
plt.tight_layout()
plt.show()
```

```python
#MONTHLY SALES AND PROFIT TRENDS
df = pd.DataFrame(data)
df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

monthly_trends = df.groupby(['Year', 'Month']).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

monthly_trends['Date'] = pd.to_datetime(monthly_trends[['Year', 'Month']].assign(DAY=1))

plt.figure(figsize=(12, 6))

plt.plot(monthly_trends['Date'], monthly_trends['Sales'], marker='o', color='blue', label='Sales')
plt.plot(monthly_trends['Date'], monthly_trends['Profit'], marker='o', color='green', label='Profit')

plt.title('Monthly Sales and Profit Trends')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
#BAR CHART FOR SEGMENT AND SALES
sales_seg = data.groupby('segment')['sales'].sum().sort_values(ascending=False)

title_font = {'family': 'calibri', 'color': 'darkblue', 'weight': 'bold', 'size': 20 }
label_font = {'family': 'serif', 'color': 'darkred', 'weight': 'normal', 'size': 14 }
tick_font = {'family': 'sans-serif', 'color': 'black', 'size': 12 }

plt.figure(figsize=[12, 8])
sns.barplot(x=sales_seg.index, y=sales_seg.values, palette='viridis')

for i, value in enumerate(sales_seg.values):
    plt.text(i, value + 0.05 * max(sales_seg.values), 
             f'${value:,.0f}', ha='center', va='bottom', fontsize=12, color='black')

plt.xlabel('Segment', **label_font)
plt.ylabel('Sales', **label_font)
plt.title('Sales Contribution by Segment', **title_font)

plt.xticks(rotation=45, ha='right', **tick_font)
plt.yticks(**tick_font)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
